```python
import torch
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

%matplotlib inline
```

```python
model = torch.hub.load("ultralytics/yolov5", "custom", path="erf.pt", _verbose=False)
```

```python
model.conf = 0.3
```

```python
# img = "https://ultralytics.com/images/zidane.jpg"
# img = "https://isefireproducts.co.uk/fire-extinguisher-colours-a-simple-guide/fire-extinguishers.webp"
img = "https://cdn.cadcrowd.com/3d-models/9a/07/9a07c126-3a34-4bb7-bf94-f30b4dd003b7/gallery/1543e35c-9ebb-412e-a423-79834e212668/medium.jpg"
```

```python
objs = model(img)
```

```python
plt.imshow(objs.render()[0])
plt.show()
```

```python
classes = list(set(objs.pred[0][...,-1].detach().cpu().tolist()))
classes
```

```python
objs.names[classes[0]]
```
